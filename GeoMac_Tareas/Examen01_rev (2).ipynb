```
#Derivadas
import numpy as np
import matplotlib.pyplot as plt

# Función a ser evaluada
y = lambda x: 1 - np.exp(x) + (np.exp(1) - 1)*np.sin(np.pi*x/2)

# Dominio de la función
x = np.linspace(-2, 1, 100)

# Gráfica de la función
plt.plot(x,y(x))
plt.grid()
plt.show()

print("La función se hace 0 en -1 y 0")
```

```
# Función a ser evaluada
y1 = lambda x: (x-1)*np.tan(x)+ x * np.sin(np.pi * x)

# Dominio de la función
x = np.linspace(-0, 1, 100)

# Gráfica de la función
plt.plot(x,y1(x))
plt.grid()
plt.show()

print("La función se hace 0 en 0 y 0.3")
```

```
# Función a ser evaluada
y2 = lambda x: x * np.sin(np.pi * x)-(x-2)* np.log(x)

# Dominio de la función
x = np.linspace(-1, 2, 100)

# Gráfica de la función
plt.plot(x,y2(x))
plt.grid()
plt.show()

print("La función se hace 0 entre [0, 1]")
```

```
# Función a ser evaluada
y3 = lambda x: (x-2)*np.sin(x)*np.log(x+2)

# Dominio de la función
x = np.linspace(-2, 2, 100)

# Gráfica de la función
plt.plot(x,y3(x))
plt.grid()
plt.show()

print("La función se hace 0 en [1, 2]")
```
