```
import numpy as np
import math
#Calcular el error absoluto y relativo

Ea = np.fabs(np.pi - 22/7)               
print ('Error absoluto a', Ea)
Er = Ea / np.fabs(np.pi) 
print ('Error relativo a', Er)         

p = math.exp(10)
pa = 22000
Ea1 = np.fabs(p-pa)               
print ('error absoluto b', Ea1)
Er1 = Ea1 / p
print ('Error relativo b', Er1) 


p = 10**np.pi
pa = 1400
Ea2 = np.fabs(p-pa)               
print ('error absoluto c',Ea2)
Er2 = Ea2 / np.fabs(p) 
print ('Error relativo c',Er2)   

Ea3 = np.fabs(math.factorial(9) - ((np.sqrt(18*np.pi)) *(9/np.exp(1))**9))
print ('error absoluto d',Ea3)
Er3 = Ea3 / np.fabs(math.factorial(9)) 
print ('Error relativo d',Er3)
```

```
#Taylor
import math
def aprox_a(x,n):
  #a = []
  a = 0 
  for i in range(0, n+1):
    a += ((-1**i)*(5**i)) / math.factorial(i)
    #a.append(x)
  return a

#print(sum(a))

def aprox_b(x,n):
  #b=[]
  b = 0
  for i in range (0, n+1):
    b += math.factorial(i) / (5**i)  
    #b.append(y)
  return b
#print(sum(b))
```

```
#Taylor 2.0

x = -5
n= 20
ea = aprox_a(x,n)
eb=  aprox_b(x,n)

print('e = {:.5f}'.format(np.exp(-5)))
print('e = {:.5f}'.format(ea))
print('e = {:.5f}'.format(eb))

#print("Lo intenté pero no sé porque no salió si ya esta igual al que puso usted en las respuestas ")
#print("intenté hacerlo con lo que ya había hecho para el examen, pero no me dejaba")
#print("hice una mezcla rara entre su código y el mío y tampoco")
#print("lo pasé a spyder porque pensé que era por la memoria del colab pero no")
#print("no sé porque me imprime esos valores, si me puede ayudar a entenderlo se lo agradecería")
#print("Más allá de que puede ayudarme este ejercicio a subir un poco la calificación, quiero entender porque me imprime eso")
#print("Saludos cordiales :3")
```

```
#Derivadas
import numpy as np
import matplotlib.pyplot as plt

# Función a ser evaluada
y = lambda x: 1 - np.exp(x) + (np.exp(1) - 1)*np.sin(np.pi*x/2)

# Dominio de la función
x = np.linspace(-2, 1, 100)

# Gráfica de la función
plt.plot(x,y(x))
plt.grid()
plt.show()

print("La función se hace 0 en -1 y 0")
```

```
# Función a ser evaluada
y1 = lambda x: (x-1)*np.tan(x)+ x * np.sin(np.pi * x)

# Dominio de la función
x = np.linspace(-0, 1, 100)

# Gráfica de la función
plt.plot(x,y1(x))
plt.grid()
plt.show()

print("La función se hace 0 en 0 y 0.3")
```

```
# Función a ser evaluada
y2 = lambda x: x * np.sin(np.pi * x)-(x-2)* np.log(x)

# Dominio de la función
x = np.linspace(-1, 2, 100)

# Gráfica de la función
plt.plot(x,y2(x))
plt.grid()
plt.show()

print("La función se hace 0 entre [0, 1]")
```

```
# Función a ser evaluada
y3 = lambda x: (x-2)*np.sin(x)*np.log(x+2)

# Dominio de la función
x = np.linspace(-2, 2, 100)

# Gráfica de la función
plt.plot(x,y3(x))
plt.grid()
plt.show()

print("La función se hace 0 en [1, 2]")
```
